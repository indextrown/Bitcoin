## 기본 선 그래프

```python
import matplotlib.pyplot as plt

# 데이터
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# 그래프 크기 설정 (예: 가로 5인치, 세로 2.5인치)
plt.figure(figsize=(5, 2.5))

# 그래프 생성
plt.plot(x, y) # 선 그래프
plt.title("Line Graph") # 제목
plt.xlabel("X-axis") # x축 이름
plt.ylabel("Y-axis") # y축 이름
plt.show()
```

## 스타일 변경

```python
plt.figure(figsize=(5, 2.5))

# 스타일 적용: 점선(--), 빨간색(r), 원형 마커(o)
plt.plot(x, y, linestyle='--', color='r', marker='o')
plt.title("Customized Line Graph")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [15, 25, 35, 45, 55]

# 두 개의 선 추가
plt.plot(x, y1, label="Dataset 1", linestyle='--', color='b')
plt.plot(x, y2, label="Dataset 2", linestyle='-', color='g')
plt.legend()  # 범례 추가
plt.title("Multiple Line Graphs")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
